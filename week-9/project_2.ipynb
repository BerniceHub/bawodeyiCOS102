```python
import tkinter as tk
from tkinter import messagebox

class CalculatePrice():

    def __init__ (self, location, weight):
        self.location = location
        self.weight = weight

    def calculate_price(self):
        if self.location.lower().strip() == "pau":
            if self.weight >= 10:
                return 2000
            else:
                return 1500
        elif self.location.lower().strip() == "epe":
            if self.weight >= 10:
                return 5000
            else:
                return 4000
        else:
            return None

def calculate():
    location = location_entry.get()
    try:
        weight = float(weight_entry.get())
        calculate = CalculatePrice(location, weight)
        price = calculate.calculate_price()
        if price is not None:
            result_label.config(text=f"The price for the delivery to {location} is {price}")
        else:
            result_label.config(text="Invalid location! We only deliver to PAU and Epe")
    except:
        messagebox.showerror("Invalid input for weight")

root = tk.Tk()
root.title("Delivery Price Calculator")
root.geometry("400x200")

tk.Label(root, text="Enter Location(PAU or Epe)").pack(pady=3)
location_entry = tk.Entry(root)
location_entry.pack()

tk.Label(root, text="Enter Weight(in kg)").pack(pady=8)
weight_entry = tk.Entry(root)
weight_entry.pack()

tk.Button(root, text="Calculate Price", command=calculate).pack(pady=17)

result_label = tk.Label(root, text="", font=('Arial',12))
result_label.pack()

root.mainloop()



```
